```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fx(x):
  return 3*x**2 - 3*x + 4

def derivative(x):
  return 6*x - 3
```

```python
x = np.linspace(-2,2,2222)
plt.plot(x,fx(x),x,derivative(x))
plt.legend(['Function', 'Derivative'])
plt.xlabel('x')
plt.ylabel('f(x), df(x)/dx')
plt.grid('on')
plt.xlim(x[[0,-1]])
plt.ylim( ( np.min([np.min(fx(x)), np.min(derivative(x))]),np.max([np.max(fx(x)), np.max(derivative(x))]) ) )
plt.show()
```

```python
localmin = np.random.choice(x)
# grad descent params 
learningRate = .01
trainingEpoch = 100

# training
for i in range(trainingEpoch):
  grad = derivative(localmin)
  localmin -= learningRate * grad

plt.plot(x,fx(x),x,derivative(x))
plt.plot(localmin,fx(localmin),'ro')
plt.plot(localmin,derivative(localmin),'ro')
plt.legend(['Function', 'Derivative', 'local min'])
plt.xlabel('x')
plt.ylabel('f(x), df(x)/dx')
plt.grid('on')
plt.xlim(x[[0,-1]])
plt.title(f'Imperical local min is {round(localmin,6)}')
plt.ylim( ( np.min([np.min(fx(x)), np.min(derivative(x))]),np.max([np.max(fx(x)), np.max(derivative(x))]) ) )
plt.show()
```
