```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
populationSize = 1_000_000
skip = 1_000

sampleSize = 100
numberOfSamples = 500

population = 1 / np.logspace(np.log10(0.001), np.log10(10), populationSize)
trueMean = np.mean(population)

np.random.shuffle(population)

plt.plot(population[::skip], 'o')
plt.xlabel('sample')
plt.ylabel('value')
plt.show()
```

```python
sampleMeans = np.zeros(numberOfSamples)
for exp in range(numberOfSamples):
  sampleMeans[exp] = np.mean(np.random.choice(population, size=sampleSize))

cumulativeAverage = np.zeros(numberOfSamples)
for i in range(1, numberOfSamples):
  cumulativeAverage[i] = np.mean(sampleMeans[:i])

cumulativeAverage2 = np.cumsum(sampleMeans) / np.arange(1, numberOfSamples + 1)
```

```python
plt.plot(sampleMeans, 'ko', label='sample means')
plt.plot([1, numberOfSamples], [trueMean, trueMean], 'r', lineWidth=5, label='true mean')
plt.plot(cumulativeAverage, 'x', label='cumulative average')

plt.xlabel('Sample number')
plt.ylabel('Mean value')
plt.legend()
plt.show()
```
