```python
from google.colab import files
import re
```

```python
file = files.upload()
```

```python
file
```

```python
text = str(file[list(file.keys())[0]])
```

```python
type(text)
```

```python
regex = re.compile(r'\d\d:\d\d.\d\d\d --> \d\d:\d\d.\d\d\d')
```

```python
regex.findall(text)
```

```python
textWithoutTimestamps = regex.sub(' ', text)
textWithoutTimestamps
```

```python
lineBreakRegex = re.compile(r'\\n\d{0,3}')
formatted = lineBreakRegex.sub('', textWithoutTimestamps)
formatted
```

```python
formattedNew = formatted[9:-1]
formattedNew
```

```python
words = formattedNew.split()
len(words)
```

```python
def censor(word):
  if not len(word) == 4:
    return word
  return word[0] + '*' * 3
newWords = [censor(word) for word in words]
```

```python
newText = ''
for word in newWords:
  newText = newText + word + ' '
newText
```

```python
with open('clear_transcript.txt', 'w') as f:
  f.write(newText)

files.download('clear_transcript.txt')
```
