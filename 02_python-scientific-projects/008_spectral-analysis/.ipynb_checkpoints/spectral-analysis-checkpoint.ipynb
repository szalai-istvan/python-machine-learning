```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import welch
```

```python
def generateAutoRegressiveData(N, alpha, noise=np.random.randn):
    k = len(alpha)
    x = np.zeros(N)
    for i in range(k, N):
        x[i] = sum(alpha*x[i-k : i]) + noise() # This type of test data generation is called auto-regression
    return x

x = generateAutoRegressiveData(N=250, alpha=[-.6, .9])
N = len(x)

x += np.sin( np.linspace(0,10*np.pi,N) )
plt.plot(x)
plt.xlabel('time')
plt.title('AR(%s) process'%k)
plt.show()
```

```python
# implementing fourier-transform by hand
fc = np.zeros(N, dtype=complex) # this is an array of complex numbers
t = np.arange(N)/N

for f in range(N):
    # 1) create a complex sin wave
    csw = np.exp(-1j*2*np.pi*f*t)
    # 2) dot product with the signal
    fc[f] = np.dot(csw,x)

fc2 = np.fft.fft(x)
```

```python
plt.plot(np.abs(fc),label='Our fourier-transform')
plt.plot(np.abs(fc2),'o', label='Numpy fft', markerfacecolor='red')
plt.legend()
plt.show()
```

```python
hz = np.linspace(0,1,int(N/2 + 1))

plt.plot(hz, np.abs(fc)[:len(hz)],label='Our fourier-transform')
plt.plot(hz, np.abs(fc2)[:len(hz)],'o', label='Numpy fft', markerfacecolor='red')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')

plt.legend()
plt.show()
```

```python
# Zero-padding the FFT
# Zero-padding creates smoother curves for small datasets in the frequency domain

x = generateAutoRegressiveData(N=21,alpha=[-.6,.9])
plt.plot(x)
plt.show()
```

```python
paddedLengths = [21, 31, 121]
for i in range(3):
    L = paddedLengths[i]
    xFft = np.fft.fft(x, n=L)
    hz = np.linspace(0,1,int(L/2+1))
    plt.plot(hz,np.abs(xFft)[:len(hz)],'.-')
plt.legend([f'{x} point FT' for x in paddedLengths])
plt.show()
```

```python
# Welch's method

N = 2000
t = np.linspace(0, 4*np.pi, N)
sin = np.sin(t)
signal1 = np.concatenate((sin, sin))
signal2 = np.concatenate((sin, -1 * sin))

plt.plot(signal1)
plt.plot(signal2)
plt.show()
```

```python
# static FFT
static1 = np.fft.fft(signal1)
static2 = np.fft.fft(signal2)
staticHz = np.linspace(0, 1, int(len(signal1)/2 + 1))

# In the welch's method we fourier-transform segments of the data. The segments may overlap
dynamicHz1,dynamicX1 = welch(signal1,nfft=N)
dynamicHz2,dynamicX2 = welch(signal2,nfft=N)

fig,ax = plt.subplots(1,2,figsize=(13,4))
ax[0].plot(staticHz, np.abs(static1)[:len(staticHz)], '-o', label='Signal 1')
ax[0].plot(staticHz, np.abs(static2)[:len(staticHz)], '-o', label='Signal 2')
ax[0].legend()
ax[0].set_xlim((0,.01))
ax[0].set_title('Static spectrum via FFT')

ax[1].plot(dynamicHz1, np.abs(dynamicX1)[:len(dynamicHz1)], '-o', label='Signal 1')
ax[1].plot(dynamicHz2, np.abs(dynamicX2)[:len(dynamicHz2)], '-o', label='Signal 2')
ax[1].legend()
ax[1].set_xlim((0,.01))
ax[1].set_title('Dynamic spectrum via Welch')


plt.show()
```
