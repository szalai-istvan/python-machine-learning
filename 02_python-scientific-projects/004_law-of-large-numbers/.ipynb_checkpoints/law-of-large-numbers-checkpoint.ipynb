<div style="background:#ddffff;border-left: 6px solid #2196f3;">
<h1 style="text-align:center;">Law of large numbers</h1>
<hr style="width:65%;height:2px;border-width:1;color:gray;background-color:gray;">
<p style="padding-left:20px;padding-right:20px;">As the number of experiment repetitions increase, the average of sample means will approximate the population mean better and better.</p>
</div>

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import IntProgress
from IPython.display import display
```

```python
population = 250_000
populationMean = 165
stdDeviation = 15
heights = np.random.randn(population) * stdDeviation + populationMean
heights = [round(h, 1) for h in heights]
print(heights[:10])
```

```python
plt.plot(heights[::500], 'kx')
plt.title(f'population mean is {round(np.mean(heights), 1)}')
plt.xlabel('Sample')
plt.ylabel('Height [cm]')
plt.show()
```

```python
numberOfSampling = 200
sampleSize = 100

sampleMeans = np.zeros(numberOfSampling)
for i in range(numberOfSampling):
    sample = np.random.choice(heights, sampleSize)
    sampleMeans[i] = np.mean(sample)

plt.plot(sampleMeans, 'k.')
plt.xlabel('Sample')
plt.ylabel('Sample mean')
plt.title('Sample means')
plt.show()
```

```python
averageOfSampleMeans = np.zeros(numberOfSampling)
for i in range(1, numberOfSampling):
    averageOfSampleMeans[i] = np.mean(sampleMeans[:i])

ymin = populationMean - stdDeviation / 10
ymax = populationMean + stdDeviation / 10

    
plt.plot(averageOfSampleMeans, 'kx', markersize=2)
plt.plot([0, numberOfSampling], [populationMean, populationMean], 'r-')
plt.xlabel('Number of samples')
plt.ylabel('Average of sample means')
plt.title('Average of sample means')
ax = plt.gca()
ax.set_ylim([ymin, ymax])
plt.legend(['Sample mean averages', 'Population mean'])
plt.show()
```

```python
numberOfMetaSamples = 100
cumulativeAverages = np.zeros((numberOfMetaSamples, numberOfSampling))
metaSampleMeans = np.zeros(numberOfSampling)
f = IntProgress(min=0, max=numberOfMetaSamples, description='Calculating...')
display(f)
allSampleMeans = np.zeros((numberOfMetaSamples, numberOfSampling))
for meta in range(numberOfMetaSamples):
    for sample in range(numberOfSampling):
        randomSample = np.random.choice(heights, sampleSize)
        metaSampleMeans[sample] = np.mean(randomSample)
    
    cAverage = np.cumsum(metaSampleMeans) / np.arange(1, numberOfSampling + 1)
    allSampleMeans[meta] = cAverage
    cumulativeAverages[meta,:] = (cAverage - populationMean) ** 2
    f.value = meta
```

```python
plt.plot(cumulativeAverages.T)
plt.show()
```

<div style="background:#ddffff;border-left: 6px solid #2196f3;">
<h1 style="text-align:center;">Central limit theorem</h1>
<hr style="width:65%;height:2px;border-width:1;color:gray;background-color:gray;">
<p style="padding-left:20px;padding-right:20px;">The distribution of sample means should be a Gaussian. This stands, even if the data are not Gaussian. </p>
</div>

```python
plt.hist(sampleMeans, bins='fd')
plt.xlabel('Sample mean')
plt.ylabel('Count')
plt.show()
```

```python
y,x = np.histogram(sampleMeans)
plt.plot((x[1:]+x[:-1])/2, y)
plt.show()
```

```python
for i in range(numberOfMetaSamples):
    y,x = np.histogram(allSampleMeans[i,:],bins=20)
    plt.plot((x[1:]+x[:-1])/2, y,color=[0.4,0.4,0.4])
plt.title('Unified histogram of all meta analysises')
plt.xlim([163, 167])
plt.show()
```
