```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack as fftpack
```

```python
numberOfPoints = int(1e5)
# white noise
whiteNoise = np.random.randn(numberOfPoints)
# brown noise
brownNoise = np.cumsum(whiteNoise)

plt.plot(whiteNoise, color=(1,1,0))
plt.plot(brownNoise, color=(.4,.2,0))
plt.legend(("White noise", "Brown noise"))
plt.title('White and brown noise')
plt.show()
```

```python
fig,ax = plt.subplots(1)
ax.plot(whiteNoise,color=(1,1,0))
ax.set_ylabel('White noise')

ax2 = ax.twinx()
ax2.plot(brownNoise, color=(.4,.2,0))
ax2.set_ylabel('Brown noise')

ax.set_title('White and brown noise')
plt.show()
```

```python
whiteNoiseX = np.abs(fftpack.fft(whiteNoise)) / numberOfPoints
brownNoiseX = np.abs(fftpack.fft(brownNoise)) / numberOfPoints

frequencyVector = np.linspace(0, 1, numberOfPoints//2 + 1)
```

```python
plt.plot(frequencyVector, brownNoiseX[:len(frequencyVector)], '.', color=(.4,.2,0))
plt.plot(frequencyVector, whiteNoiseX[:len(frequencyVector)], '.', color=(1,1,0))
plt.ylim([0, .05])
plt.ylabel('Amplitude (A.U.)') # A.U. -> arbitrary unit
plt.xlabel('Frequency (fraction of Nyquist frequency)')
plt.legend(['Brown noise', 'White noise'])
plt.show()
```

```python
# Creating pink noise
k = 5 # for noise level modulating
phases = 2*np.pi*np.random.rand(numberOfPoints//2 + 1) # uniformly distributed between 0 and 2pi
amplitude = (1 / (frequencyVector+.01)) + k * np.random.randn(numberOfPoints//2 + 1)**2

fourierSpect = np.zeros(numberOfPoints, dtype=complex)
fourierSpect[:(numberOfPoints//2 + 1)] = amplitude * np.exp(1j*phases) # Pink noise fourier transformed
pinkNoise = np.real(fftpack.ifft(fourierSpect))
pinkNoiseX = np.abs(fftpack.fft(pinkNoise)) / numberOfPoints
```

```python
fig,ax = plt.subplots(1,2,figsize=(12,4))
ax[0].plot(pinkNoise,color=(1,0,1))
ax[0].set_title('Pink noise in time domain')

ax[1].plot(pinkNoiseX[:len(frequencyVector)],'.',color=(1,0,1))
ax[1].set_title('Pink noise in frequency domain')

plt.show()
```

```python
# Creating blue noise
k = .2 # for noise level modulating
phases = 2*np.pi*np.random.rand(numberOfPoints//2 + 1) # uniformly distributed between 0 and 2pi
amplitude = np.linspace(1,2,numberOfPoints//2 + 1) + k * np.random.randn(numberOfPoints//2 + 1)

fourierSpect = np.zeros(numberOfPoints, dtype=complex)
fourierSpect[:(numberOfPoints//2 + 1)] = amplitude * np.exp(1j*phases) # Pink noise fourier transformed
blueNoise = np.real(fftpack.ifft(fourierSpect))
blueNoiseX = np.abs(fftpack.fft(blueNoise)) / numberOfPoints
```

```python
fig,ax = plt.subplots(1,2,figsize=(12,4))
ax[0].plot(blueNoise,color=(0,0,.7))
ax[0].set_title('Blue noise in time domain')

ax[1].plot(blueNoiseX[:len(frequencyVector)],'.',color=(0,0,.7))
ax[1].set_title('Blue noise in frequency domain')

plt.show()
```

```python
# define functions to create them

def generateWhiteNoise(amp):
    noise = amp * np.random.randn(numberOfPoints)
    return abs(fftpack.fft(noise) / numberOfPoints)

def generateBrownNoise(amp):
    noise = np.cumsum(amp * np.random.randn(numberOfPoints))
    return abs(fftpack.fft(noise) / numberOfPoints)

def generatePinkNoise(amp):
    fourierSpect = np.zeros(numberOfPoints,dtype=complex)
    amplitude = (1 / (frequencyVector+.01)) + 5 * np.random.randn(numberOfPoints//2 + 1)**2
    phases = 2*np.pi*np.random.rand(numberOfPoints//2 + 1)
    fourierSpect[:(numberOfPoints//2 + 1)] = amplitude * np.exp(1j*phases)
    noise = amp * np.real(fftpack.ifft(fourierSpect))
    return abs(fftpack.fft(noise) / numberOfPoints)

def generateBlueNoise(amp):
    fourierSpect = np.zeros(numberOfPoints,dtype=complex)
    amplitude = np.linspace(1,3,numberOfPoints//2 + 1) + .2 * np.random.randn(numberOfPoints//2 + 1)
    phases = 2*np.pi*np.random.rand(numberOfPoints//2 + 1)
    fourierSpect[:(numberOfPoints//2 + 1)] = amplitude * np.exp(1j*phases)
    noise = amp * np.real(fftpack.ifft(fourierSpect))
    return abs(fftpack.fft(noise) / numberOfPoints)
```

```python
# show them in one plot
whiteIntensity = 1
brownIntensity = 1
pinkIntensity = 50
blueIntensity = 1000

plt.plot(frequencyVector, generateWhiteNoise(whiteIntensity)[:len(frequencyVector)], color=(1,1,0))
plt.plot(frequencyVector, generateBrownNoise(brownIntensity)[:len(frequencyVector)], color=(.4,.2,0))
plt.plot(frequencyVector, generatePinkNoise(pinkIntensity)[:len(frequencyVector)], color=(1,0,1))
plt.plot(frequencyVector, generateBlueNoise(blueIntensity)[:len(frequencyVector)], color=(0,0,.7))

plt.ylim([0,.03])
plt.xlabel('Frequency (Fraction of Nyquist)')
plt.ylabel('Amplitude')
plt.title('Frequency domain plot of noise')
plt.show()
```
