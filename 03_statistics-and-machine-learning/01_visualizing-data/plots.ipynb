```python
# Note:
# Data types in statistics:
# Numerical: 
# - Interval: Continuous with meaningless zero (0 ˚C doesn't mean absence of heat) 
# - Ratio: Continuous with meaningful zero (0 € does mean absence of money)
# - Discrete: only given values (population of a country)
# Categorical: 
# - Ordinal: sortable (my top 3 favorite ice cream flavors)
# - Nominal: not sortable (genders)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Bar and errorbar plots

m = 30 # rows
n = 6 # columns
data = np.zeros((m,n))

for i in range(n):
    data[:,i] = 30*np.random.randn(m) * (2*i/(n-1)-1)**2 + (i+1)**2

# (Note: "data[:,i]": group by the comma first)
```

```python
fig,ax = plt.subplots(1,3,figsize=(8,2))
ax[0].bar(range(n), np.mean(data,axis=0))
ax[0].set_title('Bar plot')

ax[1].errorbar(range(n), np.mean(data,axis=0), np.std(data,axis=0,ddof=1), marker='s', linestyle='')
ax[1].set_title('Errorbar plot')

ax[2].bar(range(n), np.mean(data,axis=0))
ax[2].errorbar(range(n), np.mean(data,axis=0), np.std(data,axis=0,ddof=1), marker='.', linestyle='', color='black')
ax[2].set_title('Bar + errorbar plot')

plt.show()
```

```python
groupings = [1,2,4,5,6,8]
plt.bar(groupings, np.mean(data,axis=0))
plt.errorbar(groupings, np.mean(data,axis=0), np.std(data,axis=0,ddof=1), marker='.', linestyle='', color='black')
plt.xticks(groupings, [0,1,2,3,4,5])
plt.show()
```

```python
matrix = [  [2,5,4,3], 
            [1,1,8,6] ]

fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(8,8))
ax[0,0].imshow(matrix)

df = pd.DataFrame(matrix, columns=['p1','p2','p3','p4'])
df.plot(ax=ax[1,0], kind='bar')
ax[1,0].set_title('Grouping by row')

ax[0,1].imshow(np.array(matrix).T)
df.T.plot(ax=ax[1,1],kind='bar')
ax[1,1].set_title('Grouping by columns')

plt.show()
```

```python
# Box and whisker plots

m = 30 # rows
n = 6 # columns
data = np.zeros((m,n))

for i in range(n):
    data[:,i] = 30*np.random.randn(m) * (2*i/(n-1)-1)**2 + (i+1)**2
```

```python
plt.boxplot(data)
plt.show()
```

```python
sns.boxplot(data=data,orient='v')
plt.show()
```

```python
df = pd.DataFrame(data,columns=['Zero', 'One', 'Two', 'Three', 'Four', 'Five'])
sns.boxplot(data=df,orient='h')
plt.show()
```
