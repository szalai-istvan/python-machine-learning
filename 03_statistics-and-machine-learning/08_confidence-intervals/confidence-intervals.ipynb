<h1 style="text-align:center;">Confidence intervals</h1>
<hr style="width:65%;height:2px;border-width:1;color:gray;background-color:gray;">
<p style="padding-left:20px;padding-right:20px;padding-bottom:20px;">
    The confidence interval is the probability that an unknown population parameter falls within a range of values in repeated samples. Key assumption is that variance is a meaningful paramter of the dataset.<br>P(L&lt;mu&lt;U)=c<br><br>The formula for computing it is: C.I. = mu + t*(k) * s / sqrt(n)<br><br> [s] is the standard deviation, [n] is the sample size. [t*] is the t-value associated with one tail of the confidence interval.
</p>
</div>

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from matplotlib.patches import Polygon
```

```python
popN = int(1e7)
population = (4*np.random.randn(popN))**2

populationMean = np.mean(population)

fig,ax = plt.subplots(2, 1, figsize=(6, 4))

ax[0].plot(population[::1000], 'k')
ax[0].set_xlabel('Data index')
ax[0].set_ylabel('Data value')

ax[1].hist(population, bins='fd')
ax[1].set_xlabel('Data value')
ax[1].set_ylabel('Count')

plt.show()
```

```python
samplesize = 40
confidence = 95 # percent

randomSamples = np.random.randint(0, popN, samplesize)
sampleMean = np.mean(population[randomSamples])
sampleStd = np.std(population[randomSamples], ddof=1)

citmp = (1 - confidence/100)/2
confidenceInterval = sampleMean + stats.t.ppf([citmp, 1 - citmp], samplesize - 1) * sampleStd / np.sqrt(samplesize)

y = np.array([[confidenceInterval[0], 0], [confidenceInterval[1], 0], [confidenceInterval[1], 1], [confidenceInterval[0], 1]])
p = Polygon(y, facecolor='g', alpha=.3)

fig,ax = plt.subplots()
ax.plot([populationMean, populationMean], [0, 1.5], 'k:', linewidth=2)
ax.plot([sampleMean, sampleMean], [0, 1], 'r--', linewidth=3)

ax.add_patch(p)
ax.set_xlim([populationMean-30, populationMean+30])
ax.set_yticks([])
ax.set_xlabel('Data values')
ax.legend(('True mean', 'Sample mean', '%g%% CI region'%confidence))
ax.set_ylim([0, 2])
plt.show()
```

<div style="background:#ddffff;border-left: 6px solid #2196f3;">
<h1 style="text-align:center;">Computing confidence intervals by resampling (bootstrapping)</h1>
<hr style="width:65%;height:2px;border-width:1;color:gray;background-color:gray;">
<p style="padding-left:20px;padding-right:20px;padding-bottom:20px;">
    Instead of using a formula, we randomly and repeatedly sample from our dataset. <br><br>A bootstrapped sample is a sample which contains the same number of elements as our original sample, and is choosen from the sample, but repeated choosing is allowed. <br>Example: Dataset is x0, x1, x2, x3. A bootstrapped sample might be x0, x0, x3, x1.<br><br>
    Technically a bootstrapped dataset is allowed to have less elements than the original sample, but usually it is the same number of elements.<br>
    We create a lot of bootstrapped samples. We compute the distribution of the bootstrap sample means and calculate the confidence intervals on it.<br><br>This algorithm has no assumptions, works for any param (mean, std, ...), and is very useful when we have limited data.<br>On the other hand is gives slightly different result every time, can be slow in large datasets and gives false result if sample is not a good representation of dataset.
</p>
</div>

```python
popN = int(1e7)
population = (4*np.random.randn(popN))**2

populationMean = np.mean(population)

fig,ax = plt.subplots(2, 1, figsize=(6, 4))

ax[0].plot(population[::1000], 'k')
ax[0].set_xlabel('Data index')
ax[0].set_ylabel('Data value')

ax[1].hist(population, bins='fd')
ax[1].set_xlabel('Data value')
ax[1].set_ylabel('Count')

plt.show()
```

```python
sampleSize = 40
confidence = 95 # percent

randomSamples = np.random.randint(0, popN, sampleSize)
sampleData = population[randomSamples]
sampleMean = np.mean(sampleData)
sampleStd = np.std(sampleData, ddof=1)

numberOfBootstrappings = 1_000
bootMeans = np.zeros(numberOfBootstrappings)

for i in range(numberOfBootstrappings):
    bootMeans[i] = np.mean(np.random.choice(sampleData, sampleSize))

confidenceInterval = [0, 0]
confidenceInterval[0] = np.percentile(bootMeans, (100 - confidence)/2)
confidenceInterval[1] = np.percentile(bootMeans, 100-(100 - confidence)/2)
confidenceInterval
```

```python
fig,ax = plt.subplots(1, 1)

y,x = np.histogram(bootMeans, 40)
y = y/np.max(y)
x = 0.5 * (x[:-1] + x[1:])
ax.bar(x, y)

y = np.array([[confidenceInterval[0], 0], [confidenceInterval[1], 0], [confidenceInterval[1], 1], [confidenceInterval[0], 1]])
p = Polygon(y, facecolor='g', alpha=.3)
ax.add_patch(p)

ax.plot([populationMean, populationMean], [0, 1.5], 'k:', linewidth=2)
ax.plot([sampleMean, sampleMean], [0, 1], 'r--', linewidth=3)
ax.set_xlim([int(populationMean-30), int(populationMean+30)])
ax.set_yticks([])
ax.set_xlabel('Data values')
ax.legend(('%g%% CI region'%confidence, 'True mean', 'Sample mean', 'Empirical distribution'))
ax.grid()
plt.show()
```

```python
# Compare to the analytic formula
citmp = (1 - confidence/100)/2
confidenceInterval2 = sampleMean + stats.t.ppf([citmp, 1-citmp],sampleSize-1) * sampleStd/np.sqrt(sampleSize)
```

```python
print(f'Empirical confidence interval: {np.round(confidenceInterval, 4)}')
print(f'Analytic confidence interval:  {np.round(confidenceInterval2, 4)}')
```
