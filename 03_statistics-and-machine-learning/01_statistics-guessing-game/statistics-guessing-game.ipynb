```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
average1 = 40
average2 = 45

standardDeviation = 5.5

sampleSize1 = 40
sampleSize2 = 35
```

```python
data1 = np.random.randn(sampleSize1)*standardDeviation + average1
data2 = np.random.randn(sampleSize2)*standardDeviation + average2

sampleSizes = [sampleSize1, sampleSize2]

dataLimits = [np.min(np.hstack((data1, data2))), np.max(np.hstack((data1, data2)))]
```

```python
fig,ax = plt.subplots(1, 2, figsize=(6,4))

ax[0].violinplot(data1)
ax[0].plot(1 + np.random.randn(sampleSize1)/10,data1,'ko')
ax[0].set_ylim(dataLimits)
ax[0].axis('off')

ax[1].violinplot(data2)
ax[1].plot(1 + np.random.randn(sampleSize2)/10,data2,'ko')
ax[1].set_ylim(dataLimits)
ax[1].axis('off')

t,p = stats.ttest_ind(data1, data2)

sigtext = ('', ' not')
plt.title('The two groups are%s significantly different!'%(sigtext[int(p>.05)]))

plt.show()
```
