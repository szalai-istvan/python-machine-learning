```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
rows = 30
cols = 6

data = np.zeros((rows, cols))

for i in range(cols):
    data[:,i] = 30*np.random.randn(rows) * (2*i/(rows-1)-1)**2 + (i+1)**2
```

```python
fig,ax = plt.subplots(1, 3, figsize=(8, 2))

# Barplot
ax[0].bar(range(cols), np.mean(data, axis=0))
ax[0].set_title('Bar plot')

# Errorbar
ax[1].errorbar(range(cols), np.mean(data, axis=0), np.std(data, axis=0, ddof=1), marker='s', linestyle='')
ax[1].set_title('Only error bar')

# Both
ax[2].bar(range(cols), np.mean(data, axis=0))
ax[2].errorbar(range(cols), np.mean(data, axis=0), np.std(data, axis=0, ddof=1), marker='.', linestyle='', color='k')
ax[2].set_title('Bar plot and error bar')

plt.show()
```

```python
xcrossings = [1, 2, 4, 5, 6, 8]
plt.bar(xcrossings, np.mean(data, axis=0))
plt.errorbar(xcrossings, np.mean(data, axis=0), np.std(data, axis=0, ddof=1), marker='.', linestyle='', color='k')
plt.xticks(xcrossings)
plt.title('Bar plot and error bar')

plt.show()
```

```python
m = [
    [2, 5, 4, 3], 
    [1, 1, 8, 6]
]
df = pd.DataFrame(m, columns=['prop0', 'prop1', 'prop2', 'prop3'])

fig,ax = plt.subplots(nrows=2, ncols=2, figsize=(8, 8))

ax[0, 0].imshow(m)

df.plot(ax=ax[1, 0], kind='bar')
ax[1, 0].set_title('Grouping by rows')

ax[0, 1].imshow(np.array(m).T)
df.T.plot(ax=ax[1, 1], kind='bar')
ax[1, 1].set_title('Grouping by columns')

plt.show()
```

<div style="background:#ddffff;border-left: 6px solid #2196f3;">
<h1 style="text-align:center;">Box-and-whiskers plot</h1>
<hr style="width:65%;height:2px;border-width:1;color:gray;background-color:gray;">
<p style="padding-left:20px;padding-right:20px;">Also called boxplots.</p>
<p style="padding-left:20px;padding-right:20px;">The line in the middle of the box is the median (50% of data is larger 50% is smaller). </p>
<p style="padding-left:20px;padding-right:20px;">The bottom of the box is the quartile 1 (75% of data is larger 25% is smaller). </p>
<p style="padding-left:20px;padding-right:20px;">The top of the box is the quartile 3 (25% of data is larger 75% is smaller). </p>
<p style="padding-left:20px;padding-right:20px;">The ends of the whiskers are the "minimum" and "maximum". </p>
</div>

```python
m = 30
n = 6
data = np.zeros((m,n))

for i in range(n):
    data[:,i] = 30*np.random.randn(m) * (2*i/(rows-1)-1)**2 + (i+1)**2

plt.boxplot(data)
plt.show()
```

```python
# now with seaborn
sns.boxplot(data=data,orient='v')
plt.show()
```

```python
df = pd.DataFrame(data, columns=['one', 'two', 'three', 'four', 'five', 'six'])
sns.boxplot(data=df, orient='v')
plt.show()
```

<h1 style="text-align:center;">Histograms</h1>
<hr style="width:65%;height:2px;border-width:1;color:gray;background-color:gray;">
<p style="padding-left:20px;padding-right:20px;">Data organized into bins through their range.</p>
</div>

```python
n = 1000
data = np.exp(np.random.randn(n)/2) # Log-normal distribution
plt.plot(data, 's')
plt.show()
```

```python
bins = 40
plt.hist(data,bins=bins)
plt.show()
```

```python
y,x = np.histogram(data,bins=bins)
xCenters = (x[1:] + x[:-1]) / 2
plt.plot(xCenters, y)
plt.show()
```

```python
n = 1000

data1 = np.exp(np.random.randn(n)/2)
data2 = np.exp(np.random.randn(n)/10 + 2)
data3 = np.exp(np.random.randn(n)/2 + 1)
```

```python
k = 20
plt.hist(data1,bins=k)
plt.hist(data2,bins=k)
plt.hist(data3,bins=k)
plt.show()
```

```python
markers = ['s-', 'o-', '^-']

for i,data in enumerate([data1, data2, data3]):
    y,x = np.histogram(data,bins=k)
    xx = (x[:-1] + x[1:]) / 2
    y = y/sum(y)
    plt.plot(xx, y, markers[i], label=f'data{i+1}')

plt.xlabel('Value')
plt.ylabel('Probability')
plt.legend()
plt.grid()
plt.show()
```
