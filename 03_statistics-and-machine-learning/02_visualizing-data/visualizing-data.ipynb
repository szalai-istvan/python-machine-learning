```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Bar plots

rows = 30
cols = 6

data = np.zeros((rows, cols))

for i in range(cols):
    data[:,i] = 30*np.random.randn(rows) * (2*i/(rows-1)-1)**2 + (i+1)**2
```

```python
fig,ax = plt.subplots(1, 3, figsize=(8, 2))

# Barplot
ax[0].bar(range(cols), np.mean(data, axis=0))
ax[0].set_title('Bar plot')

# Errorbar
ax[1].errorbar(range(cols), np.mean(data, axis=0), np.std(data, axis=0, ddof=1), marker='s', linestyle='')
ax[1].set_title('Only error bar')

# Both
ax[2].bar(range(cols), np.mean(data, axis=0))
ax[2].errorbar(range(cols), np.mean(data, axis=0), np.std(data, axis=0, ddof=1), marker='.', linestyle='', color='k')
ax[2].set_title('Bar plot and error bar')

plt.show()
```

```python
xcrossings = [1, 2, 4, 5, 6, 8]
plt.bar(xcrossings, np.mean(data, axis=0))
plt.errorbar(xcrossings, np.mean(data, axis=0), np.std(data, axis=0, ddof=1), marker='.', linestyle='', color='k')
plt.xticks(xcrossings)
plt.title('Bar plot and error bar')

plt.show()
```

```python
m = [
    [2, 5, 4, 3], 
    [1, 1, 8, 6]
]
df = pd.DataFrame(m, columns=['prop0', 'prop1', 'prop2', 'prop3'])

fig,ax = plt.subplots(nrows=2, ncols=2, figsize=(8, 8))

ax[0, 0].imshow(m)

df.plot(ax=ax[1, 0], kind='bar')
ax[1, 0].set_title('Grouping by rows')

ax[0, 1].imshow(np.array(m).T)
df.T.plot(ax=ax[1, 1], kind='bar')
ax[1, 1].set_title('Grouping by columns')

plt.show()
```
