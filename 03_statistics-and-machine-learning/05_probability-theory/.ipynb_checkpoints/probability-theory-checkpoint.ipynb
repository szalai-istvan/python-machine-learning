<h1 style="text-align:center;">Probability</h1>
<hr style="width:65%;height:2px;border-width:1;color:gray;background-color:gray;">
<p style="padding-left:20px;padding-right:20px;padding-bottom:20px;">Probability is the numerical description of the likeliness of an event occurring or a statement being true. Probabilities are greater than or equal to 0 and less than or equal to 1. In a set of events the sums of probabilities should be 1. <br><br>Probability calculation should be done only on discreet data, and categories should be mutually exclusive.</p>
</div>

```python
import math
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# counts of the different events
c = np.array([1, 2, 4, 3])

probabilities = c / np.sum(c)
probabilities
```

```python
blue = 40
yellow = 30
orange = 20
totalMarbles = blue + yellow + orange

jar = np.hstack((1*np.ones(blue), 2*np.ones(yellow), 3*np.ones(orange)))

numberOfDraws = 2_000
drawColors = np.zeros(numberOfDraws)

for draw in range(numberOfDraws):
    randomMarbel = jar[np.random.randint(len(jar))]
    drawColors[draw] = randomMarbel

blueDraws = np.sum(drawColors==1) / numberOfDraws
yellDraws = np.sum(drawColors==2) / numberOfDraws
oranDraws = np.sum(drawColors==3) / numberOfDraws

plt.bar([1, 2, 3], [blueDraws, yellDraws, oranDraws], label="Proportions", color='g')
plt.plot([0.5, 1.5], [blue/totalMarbles, blue/totalMarbles], 'ko-')
plt.plot([1.5, 2.5], [yellow/totalMarbles, yellow/totalMarbles], 'ko-')
plt.plot([2.5, 3.5], [orange/totalMarbles, orange/totalMarbles], 'ko-', label="Probabilities")
plt.legend()
plt.grid(axis='y')
plt.ylim(0, 0.5)
plt.yticks(np.linspace(0, 10, 11)*0.05)
plt.xlim(0, 4)
plt.xticks([1, 2, 3], labels=['Blue', 'Yellow', 'Orange'])
plt.show()
```

<div style="background:#ddffff;border-left: 6px solid #2196f3;">
<h1 style="text-align:center;">Odds</h1>
<hr style="width:65%;height:2px;border-width:1;color:gray;background-color:gray;">
<p style="padding-left:20px;padding-right:20px;padding-bottom:20px;">Odds are not the same as probabilities. Odds are presented as a ratio, (e.g. 6:1). It is the ratio of the probability of an event not occurring to the probability of it occurring.<br>r = (1-p)/p<br>p = 1/(1+r)</p>
</div>

```python
def odd(prob):
    return (1 - prob) / prob

def oddStr(prob, precision=4):
    left = odd(prob)
    left = round(left, precision)
    right = 1
    while not left.is_integer():
        left *= 10
        right *= 10
    left = int(left)
    right = int(right)
    gcd = math.gcd(left, right)
    left //= gcd
    right //= gcd
    return f'{left}:{right}'

def prob(odd):
    return 1 / (1 + odd)
```

```python
# What are the odds of drawing a king in a deck of cards?
print(f'Odd of drawing a king in a deck of cards: {oddStr(4/52)}')
```

<div style="background:#ddffff;border-left: 6px solid #2196f3;">
<h1 style="text-align:center;">Probability mass vs. Probability density</h1>
<hr style="width:65%;height:2px;border-width:1;color:gray;background-color:gray;">
<p style="padding-left:20px;padding-right:20px;padding-bottom:20px;">
    <strong>Probability mass</strong> is a function that describes probabilities for a set of exclusive <strong>discrete</strong> events.<br>
    <strong>Probability density</strong> is a function that describes probabilities for a set of exclusive <strong>continuous</strong> events. This is a smooth and continuous function. We cannot calculate the probability of one value, we can calculate the probability of a range with integrals.
</p>
</div>

```python
N = 10_000
data1 = np.cumsum(np.sign(np.random.randn(N)))
data2 = np.cumsum(np.sign(np.random.randn(N)))

plt.plot(np.arange(N), data1, linewidth=2, label='Data 1')
plt.plot(np.arange(N), data2, linewidth=2, label='Data 2')
plt.legend()
plt.show()

# We discretize the continuous data
bins = 50
y, x = np.histogram(data1, bins)
y1 = y / np.sum(y)
x1 = (x[:-1] + x[1:]) / 2

y, x = np.histogram(data2, bins)
y2 = y / np.sum(y)
x2 = (x[:-1] + x[1:]) / 2

plt.plot(x1, y1, x2, y2, linewidth=3)
plt.legend(('Data 1', 'Data 2'))
plt.xlabel('Data value')
plt.ylabel('Probability')
plt.show()
```

<div style="background:#ddffff;border-left: 6px solid #2196f3;">
<p style="padding-left:20px;padding-right:20px;padding-top:20px;padding-bottom:20px;">
    The plots above are not theoretical probabilities. They are empirical estimates.
</p>
</div>

<div style="background:#ddffff;border-left: 6px solid #2196f3;">
<h1 style="text-align:center;">Cumulative distribution function</h1>
<hr style="width:65%;height:2px;border-width:1;color:gray;background-color:gray;">
<p style="padding-left:20px;padding-right:20px;padding-bottom:20px;">We got the CDF function by summing up (integrating) the probability density function. CDF always approaches 1.<br>A value in a CDF is the sum of all the PDF values left of it.</p>
</div>

```python
x = np.linspace(0, 5, 1_001)
p1 = stats.lognorm.pdf(x, 1)
c1 = stats.lognorm.cdf(x, 1)

p2 = stats.lognorm.pdf(x, 0.1)
c2 = stats.lognorm.cdf(x, 0.1)
```

```python
fig,ax = plt.subplots(2, 1, figsize=(4, 7))

ax[0].plot(x, p1/np.sum(p1), x, p2/np.sum(p2))
ax[0].set_ylabel('Probabilities')
ax[0].set_title('pdf(x)')
ax[0].legend(('p1', 'p2'))
ax[0].grid()
ax[0].set_xlim(x[0], x[-1])

ax[1].plot(x, c1, x, c2)
ax[1].set_ylabel('Probabilities')
ax[1].set_title('cdf(x)')
ax[1].legend(('c1', 'c2'))
ax[1].grid()
ax[1].set_xlim(x[0], x[-1])

plt.show()
```

```python
# Compute the cdf by hand
dx = x[1] - x[0]
c1x = np.cumsum(p1 * dx)

allc = np.hstack((c1, c1x))
maxc = np.max(allc)
minc = np.min(allc)
rangec = maxc - minc

plt.plot(x, c1, 'k-')
plt.plot(x, c1x, 'g--')
plt.legend(('stats module', 'hand calculated'))
plt.grid()
plt.xlim(x[0], x[-1])
plt.ylim(minc - rangec * 0.02, maxc + rangec * 0.02)
plt.yticks(np.linspace(0, 1, 6))
plt.show()
```

<div style="background:#ddffff;border-left: 6px solid #2196f3;">
<h1 style="text-align:center;padding-top:20px;">Sample estimate  distribution</h1>
<hr style="width:65%;height:2px;border-width:1;color:gray;background-color:gray;">
<p style="padding-left:20px;padding-right:20px;padding-bottom:20px;">
We draw random samples from the dataset (e.g.: height of giraffes). For example if we need the mean, we draw many random samples and compute the mean, and then we compute the mean of the means.
</p>
<p style="padding-left:20px;padding-right:20px;padding-bottom:20px;">
This process is generally called <strong>Monte Carlo sampling</strong>, which is randomly sampling from a population to estimate the unknown population parameter.
</p>
</div>

<div style="background:#ddffff;border-left: 6px solid #2196f3;">
<h1 style="text-align:center;padding-top:20px;">Sampling variability</h1>
<hr style="width:65%;height:2px;border-width:1;color:gray;background-color:gray;">
<p style="padding-left:20px;padding-right:20px;padding-bottom:20px;">One or a small number of measurements might not be represantative because of sampling variability. A sample's variability can come from <br><strong>natural variation</strong> (e.g.: height of people in a country), <br><strong>measurement noise</strong> (e.g.: measuring mg with a g scaled weightscale)<br><strong>complex systems</strong>, for example by ignoring factors (e.g.: measuring height but ignoring the age of the people measured)<br><strong>randomness</strong>
<br>To correct it we should take many samples. Averaging many samples will approximate the population parameters (law of large numbers). Also we compute confidence intervals around the parameters.
</p>
</div>

```python
x = np.linspace(-5, 5, 10_001)
theoreticalNormal = stats.norm.pdf(x)

numberOfSamples = 100

sampleData = np.zeros(numberOfSamples)
for i in range(numberOfSamples):
    sampleData[i] = np.random.randn()
    
plt.hist(sampleData, color='g', density=True, label='Samples')
plt.plot(x, theoreticalNormal, 'r', linewidth=3, label='Theoretical distribution')
plt.legend()
plt.grid()
plt.xlim(-5, 5)
plt.show()
```

```python
populationN = 1_000_000
population = np.random.randn(populationN)
population = population - np.mean(population)
```

```python
samplesizes = np.arange(30, 1000)
sampleMeans = np.zeros(len(samplesizes))

for i in range(len(samplesizes)):
    sampleIndexes = np.random.randint(0, populationN, samplesizes[i])
    sampleMean = np.mean(population[sampleIndexes])
    sampleMeans[i] = sampleMean

plt.plot(samplesizes, sampleMeans, 'o-')
plt.plot([np.min(samplesizes), np.max(samplesizes)], [0, 0], 'r')
plt.xlabel('Sample size')
plt.ylabel('Sample mean')
plt.title('Comparing results by changing the sample size')
plt.xlim(np.min(samplesizes), np.max(samplesizes))
plt.grid()
plt.show()
```

```python
cumulativeSums = np.cumsum(sampleMeans)

plt.plot(np.arange(0, len(cumulativeSums)), cumulativeSums, '-')
plt.plot([0, len(cumulativeSums)], [0, 0], 'r')
plt.xlabel('Number of sample means summed')
plt.ylabel('Mean of sample means')
plt.title('Result of summing sample means')
plt.xlim(0, len(cumulativeSums))
plt.grid()
plt.show()
```

<div style="background:#ddffff;border-left: 6px solid #2196f3;">
<h1 style="text-align:center;padding-top:20px;">Expected valus vs. average</h1>
<hr style="width:65%;height:2px;border-width:1;color:gray;background-color:gray;">
<p style="padding-left:20px;padding-right:20px;padding-bottom:20px;">Average is an empirical sample estimate based on a finite amount of data. Expected value is the expected average in a population. It can come from a very large sample.<br> The expected value is the middle of a distribution.</p>
    
</div>

```python
def average(x):
    return np.sum(x) / len(x)

def expectedValue(x, p):
    return np.sum(x * p)
```

```python
# Example: Illegal die. The probabilities of rolling the numbers are:
# 1 -> 1/4
# 2 -> 1/4
# 3 -> 1/8
# 4 -> 1/8
# 5 -> 1/8
# 6 -> 1/8

expectedValue(np.array([1, 2, 3, 4, 5, 6]), np.array([.25, .25, .125, .125, .125, .125]))
```

```python
# If every value comes with the same probability, average and expected value are the same
expectedValue(np.array([1, 2, 3, 4, 5, 6]), np.array([1/6, 1/6, 1/6, 1/6, 1/6, 1/6]))
```

<div style="background:#ddffff;border-left: 6px solid #2196f3;">
<h1 style="text-align:center;padding-top:20px;">Conditional probability</h1>
<hr style="width:65%;height:2px;border-width:1;color:gray;background-color:gray;">
<p style="padding-left:20px;padding-right:20px;padding-bottom:20px;">The probability of A, given B is equal to the probability is A and B cooccurring divided by the probability of B.<br>
    <br>P(A | B) = P(A  B) / P(B)<br><br>If A and B are independent, then P(A  B) = P(A) * P(B)</p>
    
</div>

```python
N = 10_000
spikeDuration = 10
spikeNumA = .01
spikeNumB = .05

spikeTimeSeriesA = np.zeros(N)
spikeTimeSeriesB = np.zeros(N)

spikesA = np.random.randint(0, N, int(N*spikeNumA))

for i in range(len(spikesA)):
    start = int( max(0, spikesA[i] - spikeDuration/2) )
    end   = int( min(N, spikesA[i] + spikeDuration/2) )
    spikeTimeSeriesA[start:end] = 1

spikesB = np.random.randint(0, N, int(N*spikeNumB))

for i in range(len(spikesB)):
    start = int( max(0, spikesB[i] - spikeDuration/2) )
    end   = int( min(N, spikesB[i] + spikeDuration/2) )
    spikeTimeSeriesB[start:end] = 1

plt.plot(spikeTimeSeriesA, 'r--', label='TS1')
plt.plot(spikeTimeSeriesB, 'g-', label='TS2')
plt.yticks([0, 1])
plt.legend()
plt.xlim(2000, 2200)
plt.
plt.show()
```

```python
probA = np.mean(spikeTimeSeriesA) # P(A)
probB = np.mean(spikeTimeSeriesB) # P(B)

probAB = np.mean(spikeTimeSeriesA + spikeTimeSeriesB == 2) # P(A  B)

pAgivenB = probAB / probB # P(A | B)
pBgivenA = probAB / probA # P(B | A)

print('------------------ Probabilities ------------------')
print(f'Probability of A and B spike couccurring = {probAB}')
print(f'probability of A spike                   = {probA}')
print(f'probability of A given B                 = {np.round(pAgivenB, 4)}')
print(f'probability of B spike                   = {probB}')
print(f'probability of B given A                 = {np.round(pBgivenA, 4)}')
print('---------------------------------------------------')
```

<div style="background:#ddffff;border-left: 6px solid #2196f3;">
<p style="padding-left:20px;padding-right:20px;padding-bottom:20px;padding-top:20px;">The probability of A | B is very close to the probability of A. Vice versa with B. This means that the two are highly independent. Let's try creating some correlation:</p>
</div>

```python
spikeTimeSeriesA = np.zeros(N)
spikeTimeSeriesB = np.zeros(N)

spikesA = np.random.randint(0, N, int(N*spikeNumA))

for i in range(len(spikesA)):
    start = int( max(0, spikesA[i] - spikeDuration/2) )
    end   = int( min(N, spikesA[i] + spikeDuration/2) )
    spikeTimeSeriesA[start:end] = 1

spikesB = np.random.randint(0, N, int(N*spikeNumB))
spikesB[:len(spikesA)] = spikesA # <---- Creating correlation. Wherever there is A spike, there will be B spike

for i in range(len(spikesB)):
    start = int( max(0, spikesB[i] - spikeDuration/2) )
    end   = int( min(N, spikesB[i] + spikeDuration/2) )
    spikeTimeSeriesB[start:end] = 1
```

```python
probA = np.mean(spikeTimeSeriesA) # P(A)
probB = np.mean(spikeTimeSeriesB) # P(B)

probAB = np.mean(spikeTimeSeriesA + spikeTimeSeriesB == 2) # P(A  B)

pAgivenB = probAB / probB # P(A | B)
pBgivenA = probAB / probA # P(B | A)

print('------------------ Probabilities ------------------')
print(f'Probability of A and B spike couccurring = {probAB}')
print(f'probability of A spike                   = {probA}')
print(f'probability of A given B                 = {np.round(pAgivenB, 4)}')
print(f'probability of B spike                   = {probB}')
print(f'probability of B given A                 = {np.round(pBgivenA, 4)}')
print('---------------------------------------------------')
```
