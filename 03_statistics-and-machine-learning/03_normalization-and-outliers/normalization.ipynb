```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

<h1 style="text-align:center;">Z-score</h1>
<hr style="width:65%;height:2px;border-width:1;color:gray;background-color:gray;">
<p>It mean centeres the data and makes the deviation one. For each value the mean is subtracted, and then divided by the deviation. This replaces data by "standard deviations from the mean".</p>
<p>Usually it should be used if the distribution is roughly Gaussian.</p>

```python
n = 1000
data = np.random.poisson(3,n)**2
mean = np.mean(data)
std = np.std(data,ddof=1)

plt.plot(data,'s',markersize=2)
plt.xlabel('Data index')
plt.ylabel('Data value')
plt.title(f'mean={np.round(mean,3)}, std={np.round(std,3)}')
plt.show()
```

```python
dataz = stats.zscore(data)
meanz = np.mean(dataz)
stdz = np.std(dataz,ddof=1)

plt.plot(dataz,'s',markersize=2)
plt.xlabel('Data index')
plt.ylabel('Data value')
plt.title(f'mean={np.round(meanz,3)}, std={np.round(stdz,3)}')
plt.show()
```

```python
plt.plot(data,dataz,'s')
plt.xlabel('Original')
plt.ylabel('Z-transformed')
plt.title( 'Correlation = %g'%(np.corrcoef(data,dataz)[0,0]) )
plt.show()
```

<h1 style="text-align:center;">Min-max scaling</h1>
<hr style="width:65%;height:2px;border-width:1;color:gray;background-color:gray;">
<p>Scales data so that min becomes 0 and max becomes 1. </p>
<p>The unit is "unity"</p>

```python
n = 100
data = np.log( np.random.rand(n) ) * 234 + 900

datamin = min(data)
datamax = max(data)

dataS = (data - datamin) / (datamax - datamin)

x = 1 + np.random.randn(n) / 20
fig,ax = plt.subplots(1,2,figsize=(10,4))
ax[0].plot(x,data,'ks')
ax[0].set_title('Raw data')
ax[1].plot(x,dataS,'ks')
ax[1].set_title('Normalized data')
plt.show()
```

<h1 style="text-align:center;">Outliers</h1>
<hr style="width:65%;height:2px;border-width:1;color:gray;background-color:gray;">
<p>Outliers are noise, extreme data, faults, etc.</p>
<p>They are produced by natural variation, noisy environment, faulty sensors, human error, non-cooperative research participants, etc</p>

<ul>
    <li>Z-score outlier removal is converting to Z normalized data, and data is an outlier if it exceeds some threshold. Often 3. We can repeat this until there are no outliers. This method should only be used on Gaussian data. </li>
    <li>Modified Z-score is the same method, but with different Z-score calculation.</li>
</ul>

```python
# Z-score outlier removal

n = 1000
threshold = 2.2
data = 2*np.random.randn(n)+5

dataz = np.abs(stats.zscore(data))
data_ = data
iterations = 0
while sum(dataz>threshold)>0:
    iterations+=1
    data_ = data_[dataz<threshold]
    dataz = np.abs(stats.zscore(data_))

fig,ax = plt.subplots(1,2,figsize=(10,5))
ax[0].plot(data,'ro')
ax[0].set_title(f'Unfiltered, {len(data)} datapoints')

ax[1].plot(data_,'ro')
ax[1].set_title(f'Filtered, {len(data_)} datapoints, filtered in {iterations} iterations')

ax[1].set_xlim(ax[0].get_xlim())
ax[1].set_ylim(ax[0].get_ylim())

plt.show()
```
