```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

<h1 style="text-align:center;">Z-score</h1>
<hr style="width:65%;height:2px;border-width:1;color:gray;background-color:gray;">
<p>It mean centeres the data and makes the deviation one. For each value the mean is subtracted, and then divided by the deviation. This replaces data by "standard deviations from the mean".</p>
<p>Usually it should be used if the distribution is roughly Gaussian.</p>

```python
n = 1000
data = np.random.poisson(3,n)**2
mean = np.mean(data)
std = np.std(data,ddof=1)

plt.plot(data,'s',markersize=2)
plt.xlabel('Data index')
plt.ylabel('Data value')
plt.title(f'mean={np.round(mean,3)}, std={np.round(std,3)}')
plt.show()
```

```python
dataz = stats.zscore(data)
meanz = np.mean(dataz)
stdz = np.std(dataz,ddof=1)

plt.plot(dataz,'s',markersize=2)
plt.xlabel('Data index')
plt.ylabel('Data value')
plt.title(f'mean={np.round(meanz,3)}, std={np.round(stdz,3)}')
plt.show()
```

```python
plt.plot(data,dataz,'s')
plt.xlabel('Original')
plt.ylabel('Z-transformed')
plt.title( 'Correlation = %g'%(np.corrcoef(data,dataz)[0,0]) )
plt.show()
```

<h1 style="text-align:center;">Min-max scaling</h1>
<hr style="width:65%;height:2px;border-width:1;color:gray;background-color:gray;">
<p>Scales data so that min becomes 0 and max becomes 1. </p>
<p>The unit is "unity"</p>

```python
n = 100
data = np.log( np.random.rand(n) ) * 234 + 900

datamin = min(data)
datamax = max(data)

dataS = (data - datamin) / (datamax - datamin)

x = 1 + np.random.randn(n) / 20
fig,ax = plt.subplots(1,2,figsize=(10,4))
ax[0].plot(x,data,'ks')
ax[0].set_title('Raw data')
ax[1].plot(x,dataS,'ks')
ax[1].set_title('Normalized data')
plt.show()
```

<h1 style="text-align:center;">Outliers</h1>
<hr style="width:65%;height:2px;border-width:1;color:gray;background-color:gray;">
<p>Outliers are noise, extreme data, faults, etc.</p>
<p>They are produced by natural variation, noisy environment, faulty sensors, human error, non-cooperative research participants, etc</p>

<ul>
    <li>Z-score outlier removal is converting to Z normalized data, and data is an outlier if it exceeds some threshold. Often 3. We can repeat this until there are no outliers. This method should only be used on Gaussian data. </li>
    <li>Modified Z-score is the same method, but with different Z-score calculation.</li>
</ul>

```python
# Z-score outlier removal

n = 1000
threshold = 2.2
data = 2*np.random.randn(n)+5

dataz = np.abs(stats.zscore(data))
data_ = data
iterations = 0
while sum(dataz>threshold)>0:
    iterations+=1
    data_ = data_[dataz<threshold]
    dataz = np.abs(stats.zscore(data_))

fig,ax = plt.subplots(1,2,figsize=(10,5))
ax[0].plot(data,'ro')
ax[0].set_title(f'Unfiltered, {len(data)} datapoints')

ax[1].plot(data_,'ro')
ax[1].set_title(f'Filtered, {len(data_)} datapoints, filtered in {iterations} iterations')

ax[1].set_xlim(ax[0].get_xlim())
ax[1].set_ylim(ax[0].get_ylim())

plt.show()
```

<h1 style="text-align:center;">Multivariate Z-score</h1>
<hr style="width:65%;height:2px;border-width:1;color:gray;background-color:gray;">
<p>We calculate the multi-dimensional mean of the dataset, convert all data to distance from maen, convert that into Z-score, and remove what's aboveour threshold.</p>

```python
N = 100

d1 = np.exp( -abs(np.random.randn(N)*3) )
d2 = np.exp( -abs(np.random.randn(N)*5) )
mean = [ np.mean(d1), np.mean(d2) ]

ds = np.zeros(N) # distance from mean
for i in range(N):
    ds[i] = np.sqrt( (d1[i] - mean[0])**2 + (d2[i] - mean[1])**2 )

z = (ds - np.mean(ds)) / np.std(ds,ddof=1)

threshold = 2.5
oidx = np.where(np.abs(z)>threshold)[0]

fig,ax = plt.subplots(1,2,figsize=(11,6))
ax[0].plot(d1,d2,'ko',markerfacecolor='k')
ax[0].set_xticks([])
ax[0].set_yticks([])
ax[0].set_xlabel('Variable x')
ax[0].set_ylabel('Variable y')
ax[0].plot(d1[oidx],d2[oidx],'x',color='r',markersize=20)

ax[0].plot(mean[0],mean[1],'kp',markerfacecolor='g',markersize=15)

ax[1].plot(z,'ko',markerfacecolor=[.7,.5,.3],markersize=12)
ax[1].set_xlabel('Data index')
ax[1].set_ylabel('Distance from mean')

ax[1].plot(oidx,z[oidx],'x',color='r',markersize=20)

plt.show()
```

<h1 style="text-align:center;">Data trimming</h1>
<hr style="width:65%;height:2px;border-width:1;color:gray;background-color:gray;">
<p>Should be used on mean-centered data. We sort the data and remove the most extreme k or k% of the data.</p>

```python
N = 50
data = np.random.randn(N,1)
data[data<-2] = - data[data<-2]**2
data[data> 2] = - data[data> 2]**2

dataMC = data - np.mean(data)

# Percent trim
trimPercent = 5
datacutoff = np.percentile(abs(dataMC),100-trimPercent)
data2cut = np.where( abs(dataMC)>datacutoff )[0]

# k trim
k2remove = 3
datasortidx = np.argsort(abs(dataMC),axis=0)[::-1]
data2cut_ = np.squeeze(datasortidx[:k2remove])

fig,ax = plt.subplots(1,1)
ax.plot(data,'k^',markerfacecolor='y',markersize=12)
ax.plot(data2cut,data[data2cut],'rx',markersize=15)
ax.plot(data2cut_,data[data2cut_],'go',markersize=15,alpha=.5)
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel('Data index')
ax.set_ylabel('Data value')
ax.legend(['All data',f'Data trimmed by {100-trimPercent}% threshold',f'Data trimmed by k={k2remove} threshold'])
plt.show()
```
