```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
sleepHours = [5, 5.5, 6, 6, 7, 7, 7.5, 8, 8.5, 9]
dollars = [47, 53, 52, 44, 39, 49, 50, 38, 43, 40]

plt.plot(sleepHours, dollars, 'ko', markerfacecolor='k')
plt.xlabel('Hours slept')
plt.ylabel('Dollars spent')
plt.xlim((4, 10))
plt.ylim((35, 55))
plt.grid()
plt.show()
```

```python
designMatrix = np.vstack((np.ones(10),sleepHours)).T

beta = np.linalg.lstsq(designMatrix, dollars, rcond=None)[0]
print(beta)
yHat = designMatrix@beta
yHat
```

```python
plt.plot(sleepHours, dollars, 'ko', markerfacecolor='k')
plt.plot(sleepHours, yHat, 's-')

for i in range(10):
    plt.plot([sleepHours[i], sleepHours[i]], [dollars[i], yHat[i]], 'm--')

plt.xlabel('Hours slept')
plt.ylabel('Dollars spent')
plt.xlim((4, 10))
plt.ylim((35, 55))
plt.grid()
plt.show()
```

```python
# with scipy:
slope,intercept,r,p,std_err = stats.linregress(sleepHours, dollars)
print(intercept, slope)
print(beta)
```
